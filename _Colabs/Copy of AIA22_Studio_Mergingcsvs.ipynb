Relevant activites for the research. Ideally, an OpenPose model would be trained exclusively with the activities that the researchers find important.

```python
activities_relevant = ['applauding',
                       'climbing',
                       'cooking', 
                       'cutting vegetables', 
                       'drinking', 
                       'fixing a bike', 
                       'fixing a car', 
                       'gardening', 
                       'playing guitar', 
                       'playing violin', 
                       'pushing a cart',
                       'reading', 
                       'riding a bike', 
                       'riding a horse',
                       'rowing a boat',
                       'running', 
                       'taking photos',
                       'throwing frisby', 
                       'walking the dog',
                       'writing on a board',
                       'writing on a book']
```

Merging and counting the activities and poses

```python
num_activities = 3

# Indices
indices = []
# Number of activities
num_act = []
# Activities' lists and numbers
act_1 = []
act_2 = []
act_3 = []
act_1_n = []
act_2_n = []
act_3_n = []
# Poses' lists and numbers
pos_1 = []
pos_2 = []
pos_3 = []
pos_1_n = []
pos_2_n = []
pos_3_n = []

for subfolder in os.listdir(folder_path):
  for filename in os.listdir(folder_path + '/' + subfolder):
    if filename.endswith('csv'):
      indices.append(subfolder)

      void_df = pd.read_csv(folder_path+'/'+subfolder+'/'+filename, sep=',')

      # ACTIVITIES TO LISTS FOR DF
      activities_dict = {}
      for act in void_df['activity']:
        if act in activities_relevant:
          if act in activities_dict.keys():
            activities_dict[act] += 1
          else:
            activities_dict[act] = 1

      num_act.append(len(activities_dict))

      if len(activities_dict) >= num_activities:
        max_activities = sorted(activities_dict, key=activities_dict.get, reverse=True)[:num_activities]
      else:
        max_activities = sorted(activities_dict, key=activities_dict.get, reverse=True)

      while len(max_activities) < num_activities:
        max_activities.append('nothing')
      
      act_1.append(max_activities[0])
      act_2.append(max_activities[1])
      act_3.append(max_activities[2])
      act_1_n.append(activities_dict[max_activities[0]])
      act_2_n.append(activities_dict[max_activities[1]])
      act_3_n.append(activities_dict[max_activities[2]])


      # POSES TO LISTS FOR DF
      poses_dict = {}
      for pose in void_df['pose']:
        if pose in poses_dict.keys():
          poses_dict[pose] += 1
        else:
          poses_dict[pose] = 1
      poses_dict['nothing'] = 0

      if len(poses_dict) >= num_activities:
        max_poses = sorted(poses_dict, key=poses_dict.get, reverse=True)[:num_activities]
      else:
        max_poses = sorted(poses_dict, key=poses_dict.get, reverse=True)

      while len(max_poses) < num_activities:
        max_poses.append('nothing')
    
      pos_1.append(max_poses[0])
      pos_2.append(max_poses[1])
      pos_3.append(max_poses[2])
      pos_1_n.append(poses_dict[max_poses[0]])
      pos_2_n.append(poses_dict[max_poses[1]])
      pos_3_n.append(poses_dict[max_poses[2]])
```

```python
combined_dict = {'num_act':num_act, 
                 'top_act':act_1, 
                 'top_act_n':act_1_n, 
                 '2_act':act_2, 
                 '2_act_n':act_2_n, 
                 '3_act':act_3, 
                 '3_act_n':act_3_n, 
                 'top_pose':pos_1, 
                 'top_pose_n':pos_1_n, 
                 '2_pose':pos_2, 
                 '2_pose_n':pos_2_n, 
                 '3_pose':pos_3, 
                 '3_pose_n':pos_3_n
                 }
```

```python
combined_df  = pd.DataFrame(combined_dict, index=indices)
```

Adding new metrics after counting activities and poses. Activity variation compares the number of activities detected in a square with the maximum number of activities detected in all the squares studied. Activeness considers the predicted poses and gives a value of 1 to the "upright" ones and of 0 to the "sitting" and "nothing" ones to express how active and lively the activities carried out in the squares are.

```python
max_num_act = combined_df['num_act'].max()
act_indicator = combined_df['num_act']/max_num_act
act_indicator = act_indicator.round(2)
```

```python
combined_df['activity_variation'] = act_indicator
```

```python
upright_bools_1 = combined_df['top_pose'] == 'upright'
upright_vals_1 = upright_bools_1*combined_df['top_pose_n']

upright_bools_2 = combined_df['2_pose'] == 'upright'
upright_vals_2 = upright_bools_2*combined_df['2_pose_n']

upright_bools_3 = combined_df['3_pose'] == 'upright'
upright_vals_3 = upright_bools_3*combined_df['3_pose_n']
```

```python
activeness = (upright_vals_1+upright_vals_2+upright_vals_3)/(combined_df['top_pose_n']+combined_df['2_pose_n']+combined_df['3_pose_n'])
activeness = activeness.round(2)

combined_df['activeness'] = activeness
```
